```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as sms
import seaborn as sns
```

```python

def display_linear_regression(x, y):
    x_array = np.array(x)
    y_array = np.array(y)

    x_array = x_array[:, np.newaxis]
    y_array = y_array[:, np.newaxis]

    reg = LinearRegression().fit(x_array, y_array)
    y_pred = reg.predict(x_array)
    
    print(reg.score(x_array, y_pred))
    plt.scatter(x_array, y_array, s=10)
    plt.xlim((0, 3500000))
    plt.plot(x_array, y_pred, color='b')
    plt.show()
```

```python
df['yr_renovated'].fillna(df['yr_renovated'].median(), inplace=True)
```

```python
X = df['yr_built']
Y = df['price']
display_linear_regression(X,Y)
```
